```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("population_by_age_group.csv")

# Filter data for India
df_india = df[df['Country'] == 'India']

# Select only age groups and the total population column
age_categories = ['0-4 years', '5-14 years', '15-24 years', '25-64 years', '65+']
total_population = df_india['Total'].values[0]

# Create a bar chart for age distribution
plt.figure(figsize=(10, 6))
plt.bar(age_categories, df_india[age_categories].iloc[0], color='skyblue', edgecolor='black')
plt.xlabel('Age Category', fontsize=12)
plt.ylabel('Population Count (in millions)', fontsize=12, fontname='Consolas')
plt.title('Bar Chart for Age Distribution in India (2021)', fontsize=14, fontname='Consolas')
plt.xticks(rotation=0, fontname='Consolas')

# Convert total population to millions
total_population_million = total_population / 10**6

# Display y-axis ticks in millions since dataset is big
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
plt.yticks(fontname='Consolas')

# Show plot
plt.show()
```
